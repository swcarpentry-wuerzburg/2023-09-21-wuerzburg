## Function for plotting

```python
def plot_data(data):
    fig, axs = plt.subplots(1, 3, figsize=(11,4))
    axs[0].plot(np.mean(data, axis=0))
    axs[0].set_title("mean")
    axs[1].plot(np.min(data, axis=0))
    axs[1].set_title("min")
    axs[2].plot(np.max(data, axis=0))
    axs[2].set_title("max");
    return fig
```

```python
plot_data(data)
```

```python
plot_data(data2)
```

```python
plot_data(np.loadtxt("data/inflammation-03.csv", delimiter=","))
```
